```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sbn
import json
import os
import sys
import pandas as pd
import pickle

root_path = os.environ['FCLA_ROOT']
data_path = os.environ['DATA_PATH']
sys.path.append(os.path.join(root_path,'scripts'))

from mergeResults import *
```

```python
def plot_all(dirname,expname,gurobi=False):
    results15 = load_results(os.path.join(data_path, "clustered", dirname, "solutions", "gurobi"), 
                       os.path.join(data_path,'clustered',dirname))
    results15_fcla = load_results(os.path.join(data_path, "clustered", dirname, "solutions", "fcla"), 
                               os.path.join(data_path,'clustered',dirname))
    results15_fcla_l0 = load_results(os.path.join(data_path, "clustered", dirname, "solutions", "fcla_l0"), 
                               os.path.join(data_path,'clustered',dirname))
    results15_hilbert = load_results(os.path.join(data_path, "clustered", dirname, "solutions", "hilbert"), 
                               os.path.join(data_path,'clustered',dirname))
    
    fig, ax = plt.subplots(figsize=(5, 4), dpi=300)
    plt.gca().set_yscale("log", nonposx='clip')
    plt.gca().set_xscale("log", nonposx='clip')
    results15_fcla = results15_fcla.sort_values(by='vcount')
    if (gurobi):
        results15 = results15.sort_values(by='vcount')
    results15_fcla_l0 = results15_fcla_l0.sort_values(by='vcount')
    results15_hilbert = results15_hilbert.sort_values(by='vcount')
    plt.plot(results15_fcla['vcount'],results15_fcla['runtime'],label="WMA")
    if (gurobi):
        plt.plot(results15['vcount'],results15['runtime'],label="Gurobi")
    plt.plot(results15_fcla_l0['vcount'],results15_fcla_l0['runtime'],label="WMA Greedy")
    plt.plot(results15_hilbert['vcount'],results15_hilbert['runtime'],label="Hilbert")
    plt.xlabel("Size of a network")
    plt.ylabel("runtime, sec")
    plt.title(expname)
    plt.legend(loc=4)
    plt.show()
    
    fig, ax = plt.subplots(figsize=(5, 4), dpi=300)
    plt.gca().set_yscale("log", nonposx='clip')
    plt.gca().set_xscale("log", nonposx='clip')
    results15_fcla = results15_fcla.sort_values(by='vcount')
    if (gurobi):
        results15 = results15.sort_values(by='vcount')
    results15_fcla_l0 = results15_fcla_l0.sort_values(by='vcount')
    results15_hilbert = results15_hilbert.sort_values(by='vcount')
    plt.plot(results15_fcla['vcount'],results15_fcla['objective'],label="WMA")
    if (gurobi):
        plt.plot(results15['vcount'],results15['objective'],label="Gurobi")
    plt.plot(results15_fcla_l0['vcount'],results15_fcla_l0['objective'],label="WMA Greedy")
    plt.plot(results15_hilbert['vcount'],results15_hilbert['objective'],label="Hilbert")
    plt.xlabel("Size of a network")
    plt.ylabel("Objective")
    plt.title(expname)
    plt.legend(loc=4)
    plt.show()
```

```python
plot_all('n_m-n0.05c-minc2m-k10',"10% Customers, 1% Facilities, capacity=20",True)
```
