```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
points = []
with open("../bin/1498732894829600676.ntw",'r') as f:
    info = f.readline().split()
    for i in range(int(info[2])):
        t = f.readline()
    for i in range(int(info[1])):
        t = f.readline().split()
        points.append((float(t[0]),float(t[1])))
```

```python
len(points)
```

```python
plt.scatter(np.array(points)[:,0],np.array(points)[:,1])
```
