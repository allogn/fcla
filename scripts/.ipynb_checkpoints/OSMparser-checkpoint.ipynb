```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
from scipy import spatial

#nodes of a graph
points = []
g_index_map = []
g_inv_index_map = {}
i = 0
for node in G.nodes_iter():
    g_index_map.append(node)
    g_inv_index_map[node] = i
    points.append((float(G.node[node]['lat']), float(G.node[node]['lon'])))
    i += 1
    
tree = spatial.KDTree(points)
```

```python
def get_node_trip(trip):
    if len(trip) < 2:
        return ([],[])
    
    trip_points = []
    for point in trip:
        trip_points.append([point[1], point[2]])
    node_list = tree.query_ball_point(trip_points, 20)
    
    node_set = set()
    for n in node_list:
        for nn in n:
            node_set.add(nn)
    node_set_list = list(node_set)
    
    # no any point falls in epsilon of any trip point
    if len(node_set_list) == 0:
        return ([],[])
    
    node_set_coords = np.array(points)[node_set_list]
    
    nbrs = NearestNeighbors(n_neighbors=1, algorithm='ball_tree').fit(trip_points)
    distances, indices = nbrs.kneighbors(node_set_coords)
    #now indices contain indices of trip points in trip array that are closest to a node
    #make dic out of that
    times_dic = {}
    for i in range(len(node_set_list)):
        times_dic[node_set_list[i]] = trip[indices[i][0]][0]
        
    subgraph = nx.DiGraph()
    #for each node in node_set (mapped to network) - find all neighbors who are in node_set and has bigger time
    for node in node_set_list:
        node_net_id = g_index_map[node]
        cur_time = times_dic[node]
        #get neighbors
        neighs = nx.all_neighbors(G,node_net_id)
        for neigh in neighs:
            #get id of the node in point_list
            neigh_node_point_id = g_inv_index_map[neigh]
            #check if it is in the covered node set
            if neigh_node_point_id in node_set:
                #check time of that node : get closest trip point id and get time
                time = times_dic[neigh_node_point_id]
                if time >= cur_time:
                    #add to subgraph
                    if not subgraph.has_edge(node_net_id, g_index_map[neigh_node_point_id]):
                        node_net_id2 = g_index_map[neigh_node_point_id]
                        subgraph.add_edge(node_net_id, node_net_id2, weight=G[node_net_id][node_net_id2]['weight'])
                        
    final_trips = []
    final_trips_len = []
    trip_times = []
    graphs = list(nx.weakly_connected_component_subgraphs(subgraph))
    for graph in graphs:
        lengths=nx.all_pairs_dijkstra_path_length(graph)
        max_l = -1
        max_node1 = 0
        max_node2 = 0
        for n1 in nx.nodes_iter(graph):
            for n2 in nx.nodes_iter(graph):
                try:
                    if max_l < lengths[n1][n2]:
                        max_l = lengths[n1][n2]
                        max_node1 = n1
                        max_node2 = n2
                except KeyError:
                    #no path
                    pass
        length = max_l
        nodes = nx.dijkstra_path(graph, max_node1, max_node2)
        if len(nodes) > 1:
            final_trips.append(nodes)
            final_trips_len.append(length)
            timing = []
            for n in nodes:
                timing.append(times_dic[g_inv_index_map[n]])
            trip_times.append(timing)
            
    if len(final_trips_len) == 0:
        return [],[]
    best = np.argsort(final_trips_len)[-1]
    result_path = final_trips[best]
    result_path_times = trip_times[best]
    
    return result_path, result_path_times
```

## my center

```python
def get_angle(p0, p1, c):
    p0c = math.sqrt((c[0]-p0[0])**2+(c[1]-p0[1])**2) 
    p1c = math.sqrt((c[0]-p1[0])**2+(c[1]-p1[1])**2)
    p0p1 = math.sqrt((p1[0]-p0[0])**2+(p1[1]-p0[1])**2)
    return math.acos((p1c*p1c+p0c*p0c-p0p1*p0p1)/(2*p1c*p0c));
```

```python
# calculate edge betweeness based on turn costs
Gturn = nx.Graph()
angle_stat = []
for node,neighs in G.adjacency_iter():
    point1 = (float(G.node[node]['lat']),float(G.node[node]['lon']))
    neighs_list = list(neighs)
    for neigh_id1 in range(len(neighs_list)):
        for neigh_id2 in range(neigh_id1+1, len(neighs_list)):
            neigh1 = neighs_list[neigh_id1]
            neigh2 = neighs_list[neigh_id2]
            point2 = (float(G.node[neigh1]['lat']),float(G.node[neigh1]['lon']))
            point3 = (float(G.node[neigh2]['lat']),float(G.node[neigh2]['lon']))
            angle_stat.append(get_angle(point2, point3, point1))
            angle = math.exp(math.pi - get_angle(point2, point3, point1))
#             angle = angle if angle < 7 else 7
            edge1 = (node,neigh1) if node < neigh1 else (neigh1, node)
            edge2 = (node,neigh2) if node < neigh2 else (neigh2, node)
            Gturn.add_edge(edge1, edge2, weight=angle)
```
